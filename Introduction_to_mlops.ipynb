# MLOps: What?

**MLOps:** spend less time for coding and focus to build business model, able for large scale for real problem, need full automation, but so do the data and modeling

DevOps and MLOps:
*   Benefits of DevOps include speed, reliability, scale, and security
*   DevOps best practice:
		* Continuous integration (CI): Saas GitHub actions, AWS code build :   qu trnh tch hp (integration) thng xuyn, nhanh chng hn khi code 
		* Continuous delivery (CD): deploying code automatically: IaC : cng nh thng xuyn cp nht phin bn mi (delivery)
		* Microservices: Flask, AWS Lambda 
		* Infrastructure as Code
		* Monitoring and instrumentation
		* Effective technical communication
		* Effective technical project management
