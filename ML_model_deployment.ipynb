# ML Model Deployment Types and Contents
1. Model-as-a-service, or live-scoring model
> ```
Real-time scoring, where one or a small number of records are scored, such as
when an ad is displayed on a website and a user session is scored by models to
decide what to display
````

2. Embedded model
> ```
Here the model is packaged into an application, which is then published.
A common example is an application that provides batch-scoring of requests.
```
> ```
Batch scoring, where whole datasets are processed using a model, such as in daily scheduled jobs.
```
